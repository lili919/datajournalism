## Go back to the second dataset - political parity

After carefully examing the first dataset, I want to have a look at the second dataset to see why China's political parity score is so low.

As is mentioned before, the political parity score is an aggregate of women’s representation across five indicators of political participation: heads of state or government, national cabinets (percentage of ministerial positions held by women), national legislatures (percentage of seats held by women in lower and upper houses of national legislatures), national legislature candidates (percentage of registered female candidates in the most recent elections to the lower and upper houses of national legislatures), and local legislatures. The score scales from 0-100 but China's score is 12.

### Basic Statistics - select rows
First, I need to take the 'China' row out of the whole data to see China's five indicators.

```python
china_parity = political_parity[political_parity['Country'].str.contains('China')]
china_parity
```

### Basic Statistics - rename columns
The name of columns in this data are too long to read, so I try to shorten the ones I will use later.

```python
china_parity = china_parity.rename(columns={'Number of elected and appointed female heads of state or governments since 1946': 'Female head of state since 1946', 'Percentage of ministerial positions held by women': 'Female ministerial positions','Percentage of seats held by women in the national legislature': 'Female seats in national legislature','Percentage of female candidates in the most recent elections to the national legislature': 'Female candidates in national legislature','Percentage of elected seats held by women in local government bodies': 'Female in local government bodies'})
china_parity
```

### Data Wrangling - aggregate
The table is in a wide format, which is not easy to see every column so I try to change it to a long format.

```python
parity_long = china_parity.melt(id_vars=['Country'], 
                          value_vars=['Female head of state since 1946', 'Female ministerial positions', 'Female seats in national legislature', 'Female candidates in national legislature', 'Female in local government bodies'],
                          value_name='parity percentage',
                          var_name = 'parity indicator')
parity_long
```

### Basic Statistics - find mean
Now I've found China's figures on the five indicators, but what is the average figure in the world?
So I try to find the mean of all the figures in all 194 countries.

```python
#world_parity = np.mean(political_parity)
world_parity = political_parity[['Number of elected and appointed female heads of state or governments since 1946', 'Percentage of ministerial positions held by women', 'Percentage of seats held by women in the national legislature', 'Percentage of female candidates in the most recent elections to the national legislature', 'Percentage of elected seats held by women in local government bodies']].mean()
world_parity
```

### Basic Statistics - convert Series to Dataframe
Since the figures are created into a series, so I try to transfer it to a Dataframe.

```python
parity_world = pd.DataFrame(world_parity)
parity_world.columns=["parity figure"]
parity_world
```

### Basic Statistics - reset index
Here the parity indicator has become an index so I use .reset_index() method to make parity indicator a column.

```python
parity_world = parity_world.reset_index()
parity_world = parity_world.rename(columns={'index': 'parity indicator'})
parity_world
```

### Basic Statistics - rename columns
The name of columns in this data are too long to read, so I try to shorten the ones I will use later.

```python
parity_world['parity indicator'] = parity_world['parity indicator'].replace({'Number of elected and appointed female heads of state or governments since 1946': 'Female head of state since 1946', 'Percentage of ministerial positions held by women': 'Female ministerial positions','Percentage of seats held by women in the national legislature': 'Female seats in national legislature','Percentage of female candidates in the most recent elections to the national legislature': 'Female candidates in national legislature','Percentage of elected seats held by women in local government bodies': 'Female in local government bodies'})
parity_world = parity_world.rename(columns={'parity figure': 'parity percentage'})
```

```python
parity_world
```

### Basic Statistics - create a new column
A 'Country' column is still needed here to merge with 'parity_long' table. So I'll need to create a new column in this table.

```python
parity_world['Country'] = 'World'
parity_world
```

### Data Wrangling - concat
Now it's time to merge the two tables! Since they have the same columns names, so .merge() is not suitable here, I use pd.concat() method.

```python
parity_global = pd.concat([parity_long,parity_world], join='inner')
parity_global.to_csv("parity_global.csv")
parity_global
```

```python
parity_global = parity_global.reset_index()
parity_global = parity_global.drop(columns=['index'])
parity_global = parity_global.round(2)
parity_global
```

## The thrid dataset - women's representation in Chinese community party, government and military
After finding out the problem of China's gender equality, which is women's empowerment, I found another dataset from US-China Economic and Security Review Commission which contains three tables. The three tables shows women's representations in Chinese community party, government and militaryFrom these three tables, I get more details of the inequality of women's empowerment in China. 

According to the GII dataset, women's empowerment is increasing slowly but steadily. However, when it comes to higher levels of leadership in these three tables, there are fewer women.

The first table is women's representation in Chinese community party leadership.

```python
women_leadership = pd.read_csv('women_leadership.csv')
women_leadership
```

### Data visualization - seaborn barplot
A horizantal bar contains leadership positions in Chinses community party and the approximate percentage of women's representation. The highest proportion is 12%.

Since Tableau can create better interactive barplots, so I will use the ones created by Tableau for my final project.

An attempt to do a horizontal bar including both total positions and female positions in Chinese community party.

```python
sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(10, 6))
sns.set_color_codes("pastel")
sns.barplot(x="Total Positions", y="Leadership Position", data=women_leadership,
            label="Total Positions", color="b", order=women_leadership['Leadership Position'])
sns.set_color_codes("muted")
sns.barplot(x="Positions Held by Women", y="Leadership Position", data=women_leadership,
            label="Positions Held by Women", color="b", order=women_leadership['Leadership Position'])
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="",
       xlabel="Women representation in Chinese community party leadership")
sns.despine(left=True, bottom=True)
```

An attempt to show the approximate percentage of women's representation in Chinese community party. This bar plot looks better than the first one, so I'll use this one for my final project.

```python
women_leadership['Approximate Percentage'] = women_leadership['Approximate Percentage'].str.replace('-','0').str.replace('%','').map(int)
```

```python
f, ax = plt.subplots(figsize=(10, 6))
women_leadership_bar = sns.barplot(y='Leadership Position', x='Approximate Percentage', data=women_leadership, 
           order=women_leadership.sort_values('Approximate Percentage')['Leadership Position'].values, 
           palette='viridis')
ax.set(ylabel="Leadership positions",
       xlabel="Women representation in Chinese community party leadership(%)")
women_leadership_bar.figure.savefig("women_leadership_cpp.png", bbox_inches='tight')
```
